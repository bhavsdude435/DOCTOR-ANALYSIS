# Visualizing min,max and median income

```python
 y=list(df.income)
 plt.boxplot(y)
 plt.show
```

```python
df.groupby(['gender','reduced']).mean()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df.corr(),cbar=True,annot=True,cmap='pink')
```

```python
#relation between income and visits
plt.figure(figsize=(5,5))
plt.scatter(x='income',y='visits',data=df)
plt.xlabel('income')
plt.ylabel('visits')
```

```python
#no of males and females affected by illness
sns.histplot(df.gender,bins=2)
```

```python
fig,ax1 = plt.subplots(figsize=(5,3))
df.groupby('gender')['reduced'].sum().nlargest(2).sort_values(ascending=False).plot(kind='bar')
```

```python
health_status=df.groupby(['age'],as_index=False)['illness'].sum().sort_values(by='illness',ascending=False).head(100)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=health_status,x='age',y='illness')
```

```python
health_status=df.groupby(['age'],as_index=False)['income'].sum().sort_values(by='income',ascending=False).head(100)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=health_status,x='age',y='income')
```

```python
sns.set(rc={'figure.figsize':(5,5)})
ax=sns.countplot(data=df,x="nchronic")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.set(rc={'figure.figsize':(5,5)})
ax=sns.countplot(data=df,x="lchronic")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sns.histplot(df.visits,bins=2)
```

```python
health_status=df.groupby(['illness'],as_index=False)['income'].sum().sort_values(by='income',ascending=False).head(100)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=health_status,x='illness',y='income')
```

```python
ax=sns.countplot(data=df,x='age',hue='gender')
for bars in ax.containers:
    ax.bar_laber(bars)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(x='health',y='age',data=df)
plt.xlabel('age')
plt.ylabel('health')
```

```python
label=['yes','no']
Y=df[df['freepoor']=='yes']
N=df[df['freepoor']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title('% of people having govt health isurance due to low income')
```

```python
label=['yes','no']
Y=df[df['private']=='yes']
N=df[df['private']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title('% of people having private health isurance')
```

```python
label=['yes','no']
Y=df[df['freerepat']=='yes']
N=df[df['freerepat']=='no']
x=[Y.shape[0],N.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label)
plt.title('% of people having govt health isurance due to oldage,disability or veteran status')
```

```python
label=['1','2','3','4','5']
Y=df[df['visits']==1]
N=df[df['visits']==2]
K=df[df['visits']==3]
V=df[df['visits']==4]
C=df[df['visits']==5]
x=[Y.shape[0],N.shape[0],K.shape[0],V.shape[0],C.shape[0]]
plt.figure(figsize=(5,5))
plt.pie(x,labels=label,autopct='%1.1f%%')
plt.title('% visits among patients')
plt.show()
```
